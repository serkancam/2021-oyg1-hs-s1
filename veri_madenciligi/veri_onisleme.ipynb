```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
# veriyi pandas ile csv dosyasından veri adında bir değişkene(dataFrame) aktaralım.
veri=pd.read_csv("kalp_rahatsizligi2.csv")
veri.head(3)
```

```python
veri.info()
```

```python
veri.rename(columns={
"gogus_agrisi_tipi":"gat",
"hareketsiz_kan_basinci":"hkb",
"serum_kolestrol":"sk",
"aclik_kan_sekeri":"aks",
"elektrokardiyografi":"ekg",
"en_yuksek_kalp_hizi":"eykh",
"anjin_bagli_egsersiz":"abe",
"st_depresyonu":"st_d",
"st_egimi":"st_e",
"buyuk_damarlar":"bds"},inplace=True)
veri.info()
```

```python
veri2=veri.drop(labels=["bds","talasemi"],axis=1)
veri2.info()
```

```python
veri3=veri.dropna(axis=0)
veri3.info()
```

```python
#veri.info()
# eksik verilerin giderilmesini sağlayan SimpleImputer sınıfından nesneler oluşturuldu.
imps=SimpleImputer(missing_values=np.nan,strategy="median")# strategy --> "mean","most_frequent"
impk=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
# eksik veri olan sütun değerleri birer değişkene alınacak
# imps ve impk nesneleri ile eksik verileri doldurulup yerine eksiksiz hazırlanmış veriler getirilecek.
veri_temiz=veri.copy()
bds=veri_temiz["bds"].values.reshape(-1,1)
tal=veri_temiz["talasemi"].values.reshape(-1,1)
veri_temiz["bds"]=imps.fit_transform(bds)
veri_temiz["talasemi"]=impk.fit_transform(tal)
veri_temiz.info()
```

```python
# aykırı veri tespiti ve sorunun giderilmesi
# serum kolestrol değerinin aykırı verilerinin tespit edilmesi yapılıp
# tespit edilen satırlar silinecek.
# veri_temiz.describe()
veri_temiz["sk"].describe()["25%"]


```

```python
sk_q1=veri_temiz["sk"].describe()["25%"]#q1
sk_q3=veri_temiz["sk"].describe()["75%"]#q3
sk_IQR=sk_q3-sk_q1#çeyrekler aralığı
alt_sinir=sk_q1-1.5*sk_IQR
ust_sinir=sk_q3+1.5*sk_IQR
print(alt_sinir)
print(ust_sinir)
```

```python
# alt ve üst sınır değerlerinin dışında olan değeleri görelim
# değer<alt_sınır veya değer>üst_sınır
aykirilar=veri_temiz[(veri_temiz["sk"]<alt_sinir)|(veri_temiz["sk"]>ust_sinir)]
aykirilar
```

```python
veri_temiz.drop(labels=aykirilar.index,axis=0,inplace=True)
```

```python
# veri_temiz=veri_temiz[(veri_temiz["sk"]>=alt_sinir)&(veri_temiz["sk"]<=ust_sinir)]
veri_temiz.info()
veri.head(3)
```

```python
veri_temiz.info()
```

```python
le=LabelEncoder()
for sutun in ["cinsiyet","gat","ekg","st_e","talasemi"]:
    veri_temiz[sutun]=le.fit_transform(y=veri_temiz[sutun])
veri_temiz.head(3)
```

```python
veri_temiz.iloc[:,:]=minmax_scale(X=veri_temiz.iloc[:,:].values,feature_range=(0,1))
veri_temiz.head(5)
```

```python
veri_temiz.to_csv("kalp_rahatsizligi_temiz.csv",index=False)
```
